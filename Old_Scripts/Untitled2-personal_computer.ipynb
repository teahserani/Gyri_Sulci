```python
import surfa 
import numpy as np
from surfa import Overlay
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
lh_mean = surfa.load_overlay('/Users/teahserani/Desktop/Gyri_sulci/FreeSurferOutputs/388_21yrOld/surf/lh.pial.H')
```

```python
lh = surfa.load_mesh('/Users/teahserani/Desktop/Gyri_sulci/FreeSurferOutputs/388_21yrOld/surf/lh.pial')
```

```python
lh_mean.data>0
```

```python
mask = lh_mean.data > 0

# Apply the mask to the vertices (broadcasting)
masked_vertices = lh.vertices * mask[:, np.newaxis]
```

```python
np.shape(masked_vertices)
```

```python
from sklearn.cluster import DBSCAN
import numpy as np

db = DBSCAN(eps=0.1, min_samples=5).fit(masked_vertices)  # Adjust `eps` and `min_samples` based on your data

# Get cluster labels
labels = db.labels_

# Visualize clusters (optional)
import matplotlib.pyplot as plt
plt.scatter(masked_vertices[:, 0], masked_vertices[:, 1], c=labels)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure

fig = plt.figure(figsize=(30, 30))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface using vertices and faces
ax.plot_trisurf(lh.vertices[:, 0], lh.vertices[:, 1], lh.vertices[:, 2], triangles=lh.faces, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Show the plot
plt.show()
```
