```python

for subject_id in subjects:
    try:
        # Path to the JSON file
        json_path = f'/Volumes/Teah2/Gyri_sulci/SG_Outputs/{subject_id}_gy-sulc_Output/FinalOutputs/averages.json'
        
        # Check if JSON exists
        if not os.path.isfile(json_path):
            print(f" JSON not found for subject: {subject_id}")
            continue

        # Load the JSON
        with open(json_path, 'r') as f:
            data = json.load(f)

        # Extract age from dataframe
        age_series = df[df['ORIG SUBJECT ID'].str.contains(subject_id, na=False)]['AGE']
        if age_series.empty:
            print(f" Age not found in DataFrame for subject: {subject_id}")
            continue

        new_key = "Age"
        new_value = int(age_series.iloc[0])

        # Insert the Age key-value pair at the second position
        items = list(data.items())
        items.insert(1, (new_key, new_value))
        new_data = dict(items)

        # Save back to the JSON
        with open(json_path, 'w') as f:
            json.dump(new_data, f, indent=4)

        print(f" Updated JSON for subject: {subject_id}")

    except Exception as e:
        print(f" Error processing {subject_id}: {e}")
```
